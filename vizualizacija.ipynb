```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np  
import matplotlib.colors as mcolors
from dataclasses import dataclass
import sympy as sp
```

DIVERGENCE PODACI

```python
raw_data = np.load('divergence_podaci.npz')
```

```python
boje = ['white', 'blue']
moj_cmap = mcolors.LinearSegmentedColormap.from_list("BelaUPlavu", boje)
divergence_ucitani_podaci = raw_data['data']
X =  raw_data['x_grid']
Y =  raw_data['y_grid']
U = raw_data['u']
V = raw_data['v']

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize = (10,3))

ax1.pcolormesh(X, Y, divergence_ucitani_podaci, cmap = moj_cmap)
ax2.quiver(X,Y,U,V)
ax3.pcolormesh(X, Y, divergence_ucitani_podaci, cmap = moj_cmap)
ax3.quiver(X,Y,U,V)
plt.show()
```

```python
vector_data = np.load('nova_brzina.npz')
```

```python
boje = ['white', 'blue']
moj_cmap = mcolors.LinearSegmentedColormap.from_list("BelaUPlavu", boje)
divergence_ucitani_podaci = vector_data['divergence']
X =  vector_data['x_grid']
Y =  vector_data['y_grid']
U = vector_data['u']
V = vector_data['v']
lap_u, lap_v = vector_data['laplacian']
fig, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(2, 3, figsize = (10,6))

ax1.pcolormesh(X, Y, divergence_ucitani_podaci, cmap = moj_cmap)
ax1.set_title('divergence')

ax2.quiver(X,Y,U,V)
ax2.set_title('vektori')

ax3.set_title('vektori + divergence')
ax3.pcolormesh(X, Y, divergence_ucitani_podaci, cmap = moj_cmap)
ax3.quiver(X,Y,U,V)

ax4.set_title('laplasijan po u')
ax4.pcolormesh(X, Y, lap_u, cmap = moj_cmap)

ax5.set_title('laplpasijan po v')
ax5.pcolormesh(X, Y, lap_v, cmap = moj_cmap)
plt.show()
```
